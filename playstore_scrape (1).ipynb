```
import re
def desc(text):
  c = 0
  l = ["mind","mindfulness","stress","anxiety","depression","mood","mental","behavioral","depress","depressed","stressed","anger","angry","frustrate","frustrated",
       "sad","lonely","moods","panic","autism","moody","bipolar","psychotic","neurology","neuro","personality","trauma","traumatic","psycho","ptsd","schizophrenia",
       "schizophrenic","dementia","dementic","psychiatric","calm","counseling","counsel","patient","patience","caregiver","empathy","behavior","minds","affirmations",
       "affirmation","headspace","therapists","therapy","therapist","self-help","emotion","emotions","self-awareness"]
  if type(text) == str:
    text = re.sub("[^a-zA-Z]", " ", text)
  textd = text.lower()
  if type(textd) == str:
    textd = textd.split(" ")
  else:
    textd = textd.decode().split(" ")
  for word in textd:
    if word in l:
      c += 1
  return c
```

```
new = df[df.description_html.map(desc)==0].index.tolist()
```
